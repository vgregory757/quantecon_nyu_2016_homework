## Exercise 1.

Use the fact that

\begin{equation}
\| A \| := \sup_{\mathbf{x}} \Big\{ \frac{\|A\mathbf{x}\|}{\|\mathbf{x}\|} : \mathbf{x} \neq \mathbf{0} \Big\} = \sup_{\mathbf{x}} \Big\{ \|A\mathbf{x}\| : \| \mathbf{x} \| = 1 \Big\}
\end{equation}

**Claim**: $\| A \| = \sqrt{\rho(A'A)}$, or equivalently $\| A \|^2 = \rho(A'A)$ for all $A \in \mathcal{M}(n \times n)$.

Use the Lagrangian for the maximization over the edge of the unit-sphere.

\begin{equation}
L(\mathbf{x}, \lambda) = \langle A\mathbf{x}, A\mathbf{x} \rangle - \lambda(\langle \mathbf{x}, \mathbf{x} \rangle - 1) = \langle A'A\mathbf{x}, \mathbf{x} \rangle - \lambda(\langle \mathbf{x}, \mathbf{x} \rangle - 1)
\end{equation}


The first-order condition with respect to $\mathbf{x}$ is:

\begin{align}
2A'A \mathbf{x} - 2 \lambda \mathbf{x} &= 0 \\
A'A \mathbf{x} &=  \lambda \mathbf{x} 
\end{align}

Hence, $\mathbf{x}$ is an eigenvector of $A'A$ and $\lambda$ is the corresponding eigenvector.

Pre-multiply by $\mathbf{x}'$ to get

\begin{align}
\mathbf{x}'A'A \mathbf{x} &=  \mathbf{x}' \lambda \mathbf{x} \\
\langle A\mathbf{x}, A\mathbf{x} \rangle &= \lambda \|\mathbf{x}\| = \lambda
\end{align}

The left hand side is the objective function $\mathbf{x}$ is maximizing---weakly positive by being a norm---hence $\lambda$ has to be the largest eigenvalue of $A'A$.

## Exercise 2.

Gelfand's formula states that for all $A \in \mathcal{M}(n \times n)$,

\begin{equation}
\| A^k \|^{\frac{1}{k}} \to \rho(A) \quad \text{as} \quad k \to \infty.
\end{equation}

**Claim:** If $\rho(A) < 1$ then there exists an $r<1$ and a $C \in \mathbb{N}$ such that $\| A^k \| \leq r^k C$.

Take $\epsilon > 0$ such that $\rho(A) + \epsilon < 1$. Then by the convergence in Gelfand's formula we have that there exists $K \in \mathbb{N}$ such that $\forall \ k \geq K$ we have that

\begin{equation}
\Big| \| A^k \|^{\frac{1}{k}} - \rho(A) \Big| < \epsilon,
\end{equation}

and in particular we have that $\forall \ k \geq K$

\begin{equation}
\| A^k \|^{\frac{1}{k}} < \rho(A) + \epsilon.
\end{equation}

Next, define $r:= \rho(A) + \epsilon < 1$. Then $\forall \ k \geq K$ we have that

\begin{equation}
\| A^k \| < r^k.
\end{equation}

For $k<K$ let's consider the maximum of the following ratio---since we have finitely many $k < K$, we can safely take the maximum.

\begin{equation}
\tilde{C} = \max_{k \in \{1, \dots, K-1\}} \frac{\| A^k \|}{r^k}
\end{equation}

Let $C:= \lfloor \max \{1, \tilde{C}\} \rfloor$.

Then, for the above defined $r<1$ and $C \in \mathbb{N}$, we have that for all $k \in \mathbb{N}$

\begin{equation}
\| A^k \| \leq r^k C.
\end{equation}

Using convergence of the sum of geometric series we immediately have $\sum_{k=0}^{\infty} \|A^k\| < \infty$.

Use the fact that for all $X \in \mathcal{M}(n \times n)$ we have that

\begin{equation}
L^k X \to X^* \quad \text{as} \quad k \to \infty
\end{equation}

Now, observe that for each $k$ we can write the Lyapunov operator as 

\begin{equation}
L^k X = A^k X (A^k)' + A^{k-1}M(A^{k-1})' + \cdots + M.
\end{equation}

As $\rho(A)<0$ we have that $\lim \|A^k\| = 0$ and hence $\lim A^k = 0$, the zero matrix.

This implies that $\lim A^k X (A^k)' = 0$ as we have that $\| A^k X (A^k)' \| \leq \| A^k\| \cdot \| X\| \cdot \|A^k\|$.

Hence, we have that $\lim L^k X$ is independent of $X$ and depends on $A$ and $M$ only.

\begin{equation}
X^* = \lim L^k X = \lim A^k X (A^k)' + \sum_{k=0}^{\infty} A^{k-1}M(A^{k-1})' = \sum_{k=0}^{\infty} A^{k-1}M(A^{k-1})'
\end{equation}

Now, we can proceed to show the claims in the exercise.

1. $M$ symmetric $\implies$ $X^*$ symmetric

    > For any $k$ we have that $(A^k M (A^k)')' = A^k M' (A^k)'$

    > Since $M$ is symmetric all the terms in the sum are symmetric

    > The sum of symmetric matrices is symmetric, hence $X^*$ is symmetric

2. $M$ non-negative definite $\implies$ $X^*$ non-negative definite

    > For any $k$ we have that $A^k M (A^k)'$ is a symmetric matrix and all its diagonal elements are non-negative as they are given by $(A^k)'_i M (A^k)_i$ and $M$ is non-negative definite.

    > As a result we have that $A^k M (A^k)'$ is non-negative definite for all $k$

    > The sum of non-negative definite matrices is non-negative definite, hence $X^*$ is non-negative definite

3. $M$ positive definite $\implies$ $X^*$ positive definite

    > The same argument as in part 2. goes through for the diagonals being strictly positive

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
class LinQ_Asset:
    """Uses an iterative procedure to solve for the unique
    solution of the Lyapunov equation in the asset pricing
    example.
    The pricing formula is given by
        p_t = b * E_t[d_{t+1} + p_{t+1}]
    Dividend:
        d_t = (x_t)'D x_t
    State Process:
        x_{t+1} = A x_t + C w_{t+1}
       
    Parameters
    ----------
    A: square matrix
    C: square matrix
    D: square matrix
    b: discount factor
    
    Returns
    -------
    A function defining equilibrium price given state vector.
    
    """
    
    
    def __init__(self, A, C, D, b):
        self.A, self.C, self.D, self.b = A, C, D, b
    
        if A.shape[0] != A.shape[1] or C.shape[0] != C.shape[1] or C.shape[0] != C.shape[1]:
            raise ValueError('The matrices have to be square.')
        if A.shape != C.shape != D.shape:
            raise ValueError('The sizes of the matrices have to be equal.')
        if b >= 1:
            raise ValueError('The discount factor has to be less than one.')

        # Test for stability condition
        if max(abs(np.linalg.eig(A)[0])) >= 1/(np.sqrt(b)):
            raise ValueError('Spectral radius of matrix is greater than unity. No unique solution exists.')
        
        # Define terms for the Lyapunov operator
        self.M = self.b * self.A.T @ self.D @ self.A
        self.L = np.sqrt(self.b) * self.A.T
        
    
    
    def lyapunov_op(self, P):
        """Defines the discrete Lyapunov operator."""
        return self.L @ P @ self.L.T + self.M
    
    
    def solv_lyapunov(self, tol=.001, maxiter=10000):
        """Iteratively solve Lyapunov equation."""
        
        # Define initial value for P
        P = np.ones(A.shape)
        
        # Set up iteration
        dist = 2 * tol
        current_iter = 0
        
        while dist > tol and current_iter < maxiter:
            P_new = self.lyapunov_op(P)
            dist = np.linalg.norm((P_new - P), 2) # use spectral norm
            P = P_new
            current_iter += 1
        return P
    
    
    def price(self, x):
        """Returns the equilibrium price."""
        P = self.solv_lyapunov()
        d = (self.b/(1 - self.b)) * np.trace(self.C.T @ (self.D + P) @ self.C)
        return x.T @ P @ x + d
```

Now, test the pricing function with the given parameter values.

```python
A = np.array([[.8, -.1], [-.1, .8]])
C = np.eye(2)
D = np.eye(2)

b = .9

# Create an instance of the LQ asset pricing class
asset = LinQ_Asset(A, C, D, b)
```

Create a meshgrid and plot the contour map of the equilibrium pricing function.

```python
grid = 100
x = np.linspace(-10.0, 10.0, grid)
y = np.linspace(-10.0, 10.0, grid)
X, Y = np.meshgrid(x, y)

Z = np.zeros((grid,grid))

for i in range(grid):
    for j in range(grid):
        Z[i][j] = asset.price(np.array([X[i][j], Y[i][j]]))
        
plt.figure(figsize = (12,8))
plt.grid(True)
plt.contourf(X, Y, Z, alpha = .4)
CS = plt.contour(X, Y, Z)
plt.clabel(CS, inline=1, fontsize=14, colors='black')
plt.title('Contour map of equilibrium pricing function', fontsize=20)
```
