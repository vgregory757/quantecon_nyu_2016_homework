### Exercise 1

We want to solve:
\begin{align}
    \max || A x || \\
    \text{s.t.} \; ||x|| = 1
\end{align}

which is equivalent to solve
\begin{align}
    \max || A x ||^2 \\
    \text{s.t.} \; ||x||^2 = 1
\end{align}

if and olny if
\begin{align}
    \max <Ax,Ax> \\
    \text{s.t.} \; <x,x> = 1
\end{align}

iff
\begin{align}
    \max x'A'Ax \\
    \text{s.t.} \; x'x = 1
\end{align}

Using the lagrange method
\begin{equation}
    \max x'A'Ax + \lambda (1-x'x)
\end{equation}

FOC:
\begin{equation}
    2A'A x - 2\lambda x = 0
\end{equation}

iff
\begin{equation}
    (A'A - \lambda I) x = 0
\end{equation}

From the last equation we can see that the optimal $x$ is an eigenvector of $A'A$ and thus $A'A x = \lambda x$

\begin{align}
    A'Ax & = \lambda x \\
    \iff x' A' A x & = \lambda x' x \\
    \iff x' A' A x = || Ax ||^2 & = \lambda || x || = \lambda
\end{align}

since we want the maximum, $\lambda$ is the largest eigenvalue, ie

\begin{equation}
    ||Ax|| = \sqrt{\rho(A'A)}
\end{equation}

### Exercise 2

Gelfand's formula $|| A^k ||^{1/k} \to \rho(A)$. Hence $|| A^k || \to \rho(A)^k$

If Hence $|| A^k || \to \rho(A)^k$, then this implies that $\forall \varepsilon > 0 \;\exists \bar{k}>0$ such that

\begin{equation}
    \bigg| \frac{||A^k||}{\rho(A)^k} -1 \bigg| \leq \varepsilon
\end{equation}

Now consider the first $\bar{k}$ terms and define:
\begin{equation}
\tilde{C} = \max \frac{|| A^k ||}{\rho(A)^k}
\end{equation}

By definition, $||A^k|| \leq C \rho(A)^k$, for all $k \in \{0, 1, \ldots, \bar{k}\}$

For all $k > \bar{k}$, we know that $\frac{||A^k||}{\rho(A)^k} \leq (1+\varepsilon)$

Define $C = \max(\tilde{C},(1+\varepsilon))$ and we obtain that for all $k \in \mathbb{N}$
\begin{equation}
    || A^k || \leq C \rho(A)^k = C r^k
\end{equation}

with $r<1$.

Finally, to show that $\sum_{k=0}^{\infty} || A^k || < \infty$ is straightforward. In fact,

\begin{equation}
    \sum_{k=0}^{\infty} || A^k || \leq \sum_{k=0}^{\infty} \rho(A)^k C = \frac{C}{1-\rho(A)} < \infty
\end{equation}

It is enoguh to show that every convergent sequence of nonnegative definite matrices converges to a nonnegative matrix. Let $\{A_n\}_{n \in \mathbb{N}}$ be a sequence of nonnegative definite matrices with limit $A$. I need to show that $A$ is nonnegative definite, ie $x'Ax \geq 0 \; \forall x \in \mathbb{R}^n \backslash 0$. In fact,

\begin{align}
    x'A x & = x'(A_n-A)x + x'A x \\
    & \geq x'(A_n-A)x
\end{align}

Taking the limit when $n \to \infty$ we obtain
\begin{equation}
    x'A x \geq 0
\end{equation}

### Exercise 5

1.- M symmetric $\Rightarrow$ $X^*$ symmetric.
  
\begin{align}
        X^* &= A X^* A' + M \\
        {X^*}' &= {A'}' {X^*}' A' + M' \qquad \text{(transpose)} \\ 
        {X^*}' &= A {X^*}' A' + M  \qquad \text{($M$ symmetric)} \\
\end{align}

Hence, ${X^*}'$ is also a solution of the Lyapunov equation. Since we know that there is a unique solution, then ${X^*}' = X^*$, so the matrix is symmetric. 

2.- M nonnegative definite $\Rightarrow$ $X^*$ nonnegative definite.

Consider the operator $LX = AXA' + M$. Since $\mathcal{M}(n \times n)$ is a complete metric space, and we proved (exercise 4) that the set of nonnegative definite matrices is closed, then it is enough to show that for a given $X$ nonnegative definite then $LX$ is also nonnegative definite. This defines a convergent sequence that converges (in the space of nonnegative definite matrices) to $X^*$.  

A matrix X is nonnegative definite if $\forall y \in \mathbb{R}^n \backslash 0$ then $y'Xy \geq 0$. Let $y \in \mathbb{R}^n \backslash 0$. Then

\begin{align}
    y'(LX)y &= y'(AXA' + M)y \\
    & = y'AX A'y + y'M y \\
    & = (A'y)'X (A'y) + y'M y
\end{align}

Note that $A'y \in \mathbb{R}^n \backslash 0$. Since $X$ is nonnegative definite the $(A'y)'X (A'y) \geq 0$. By hypothesis $M$ is nonnegative definite and therefore $y'M y \geq 0$. We conclude then that $y'(LX)y \geq 0$ and thus $LX$ is nonnegative definite.

3.- M positive definite $\Rightarrow$ $X^*$ positive definite.

A matrix X is positive definite if $\forall y \in \mathbb{R}^n \backslash 0$ then $y'Xy > 0$. Let $y \in \mathbb{R}^n\backslash 0$. Then I have to show that $y' X^* y > 0$

\begin{align}
    y'X^*y &= y'(AX^*A' + M)y \\
    & = y'AX^* A'y + y'M^*y \\
    & = (A'y)'X^* (A'y) + y'M y
\end{align}

Note that since $M$ is positive definite, the last term is positive $y' M y > 0$. Furthermore, since $M$ is positive definite, it is in general nonnegative definite, and from the previous part we know then than $X^*$ is at least nonnegative definite, and hence $(A'y)'X^* (A'y) \geq 0$. Summing up these two terms we conclude that $$y' X^* y >0$$
Therefore $X^*$ is positive definite.

In this question I create a function that takes the primitives of the linear quadratic pricing model and returns the price.

```julia
"""
This function solves the Lyapunov equation.
It finds the solution of X = AXA' + M
It solves this equation iteratively. 
"""
function Lyapunov(A, M, tol=1e-6)
    if maximum(eigvals(A)) >=1
        throw(ErrorException("Maximum eigenvalue must be < 1"))
    end
    
    x = eye(size(A,1))
    d = 1
    while d>tol 
        Lx = A*x*A'+M
        d = norm(Lx-x,Inf)
        x = copy(Lx)
    end
    
    return x
end    
```

```julia
"""
This function solves the linear quadratic asset pricing model. It returns a matrix and a constant that define
the pricing function. For a given state x, the price is p(x) = x'*P*x + δ
Inputs:
β: discount factor  
D: n by n matrix, dividends follow d = x'*D*x  
A: n by n matrix, state vector follows x = A*x + C*w, E[w*w'] = I  
C: n by n matrix.  
Outpus:  
P: matrix
δ: constant
"""
function quadratic_pricing_model(β,D,A,C)
    Λ = sqrt(β)*A'
    M = β*A'*D*A
    P = Lyapunov(Λ,M)
    #P = solve_discrete_lyapunov(Λ,M)  # if want to use QuantEcon
    δ = trace(C'*(D+P)*C)*β/(1-β)
    
    return P, δ
    
end
```

Tets the code according parameters given.

```julia
A = [.8 -0.1; -0.1 0.8]
C = eye(2)
D = eye(2)
β = 0.9
```

```julia
P, δ = quadratic_pricing_model(β,D,A,C)
```
